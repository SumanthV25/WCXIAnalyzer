```python
bowl= pd.concat([keep1,keep2], axis=0, ignore_index=True)
bowl.info()
```

```python
bowl['catches']= bowl['catches'].replace(to_replace='-',value='0')
bowl['stumpings']= bowl['stumpings'].replace(to_replace='-',value='0')
bowl['dismissals']= bowl['dismissals'].replace(to_replace='-',value='0')
bowl['dismissals/innings']= bowl['dismissals/innings'].replace(to_replace='-',value='0')

bowl
```

```python
b = pd.merge(keep1,keep2, on='Player',how='outer' )
b
```

```python
b['catches_x']= b['catches_x'].replace(to_replace='-',value='0')
b['catches_y']= b['catches_y'].replace(to_replace='-',value='0')
b['stumpings_x']= b['stumpings_x'].replace(to_replace='-',value='0')
b['stumpings_y']= b['stumpings_y'].replace(to_replace='-',value='0')
b['dismissals_x']= b['dismissals_x'].replace(to_replace='-',value='0')
b['dismissals_y']= b['dismissals_y'].replace(to_replace='-',value='0')
b['dismissals/innings_x']= b['dismissals/innings_x'].replace(to_replace='-',value='0')
b['dismissals/innings_y']= b['dismissals/innings_y'].replace(to_replace='-',value='0')
```

```python
b= b.fillna(0)
b['catches_x'] = b['catches_x'].astype(int)
b['catches_y'] = b['catches_y'].astype(int)
b['stumpings_x'] = b['stumpings_x'].astype(int)
b['stumpings_y'] = b['stumpings_y'].astype(int)
b['dismissals_x']=b['dismissals_x'].astype(int)
b['dismissals_y']=b['dismissals_y'].astype(int)
```

```python
if '2022-2022' in b['span_y'].values:
    b.loc[b['span_y'] == '2022-2022', 'span_x'] = b['span_x'].replace('2021-2021', '2021-2022')
b['matches'] = b['matches_x'] + b['matches_y']
b['innings'] = b['innings_x'] + b['innings_y']
b['dismissals']= b['dismissals_x']+b['dismissals_y']
b['catches']= b['catches_x']+b['catches_y']
b['stumpings']= b['stumpings_x']+b['stumpings_y']
b['dismissals/innings']= b['dismissals']/b['innings']


b.rename(columns = {'span_x':'span'}, inplace = True)
b['span'] = b['span'].mask(b['span'] == 0, b['span_y'])
b.head()
```

```python
b=b.drop(['dismissals/innings_y','max dismissals in innings_y','stumpings_y','stumpings_x','catches_y','catches_x','dismissals_x','dismissals_y','max dismissals in innings_x','dismissals/innings_x','innings_x','innings_y','matches_x','span_y','matches_y', ],axis=1)
b
```
