```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("C:/python projects/csv_files - Copy/batting_records.csv")
df1.head()
```

```python
df2 = pd.read_csv("C:/python projects/csv_files - Copy/bowling_records.csv")
df2.head()
```

```python
df3 = pd.read_csv("C:/python projects/csv_files - Copy/keeping_records.csv")
df3.head()
```

```python
b = pd.merge(df1,df2, on='Player',how='outer' )
b
```

```python
b.info()
```

```python
c = pd.merge(b,df3, on='Player',how='outer' )
c
```

```python
c.info()
```

```python
# c.to_csv('playing_records.csv', index=False)
```

```python
c= c.fillna(0)
```

```python
c['span_x'] = c['span_x'].mask(c['span_x'] == 0, c['span_y'])
```

```python
c.rename(columns = {'span_x':'Year'}, inplace = True)
c['Year'] = c['Year'].mask(c['Year'] == 0, c['span'])
```

```python
c
```

```python
c['matches_x'] = c['matches_x'].mask(c['matches_x'] == 0, c['matches_y'])
c['matches_x'] = c['matches_x'].mask(c['matches_x']<c['matches_y'], c['matches_y'])
c['matches_x'] = c['matches_x'].mask(c['matches_x'] == 0, c['matches'])
```

```python
c
```

```python
c['innings_x'] = c['innings_x'].mask(c['innings_x'] == 0, c['innings_y'])
c['innings_x'] = c['innings_x'].mask(c['innings_x']<c['innings_y'], c['innings_y'])
c['innings_x'] = c['innings_x'].mask(c['innings_x'] == 0, c['innings'])
```

```python
c = c.drop(['matches','matches_y','innings','innings_y','span_y','span' ], axis=1)
```
