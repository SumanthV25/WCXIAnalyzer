```python
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
bowl1= pd.read_csv("C:/python projects/scraping csv files/bowling_records1.csv")
bowl1.head()
```

```python
bowl2= pd.read_csv("C:/python projects/scraping csv files/bowling_records2.csv")
bowl2.head()
```

```python
bowl= pd.concat([bowl1,bowl2], axis=0, ignore_index=True)
bowl.info()
```

```python
bowl['maidens']= bowl['maidens'].replace(to_replace='-',value='0')
bowl['four wicket haul']= bowl['four wicket haul'].replace(to_replace='-',value='0')
bowl['fifer']= bowl['fifer'].replace(to_replace='-',value='0')

bowl.head()
```

```python
b = pd.merge(bowl1,bowl2, on='Player',how='outer' )
b
```

```python
b['maidens_x']= b['maidens_x'].replace(to_replace='-',value='0')
b['maidens_y']= b['maidens_y'].replace(to_replace='-',value='0')
b['four wicket haul_x']= b['four wicket haul_x'].replace(to_replace='-',value='0')
b['four wicket haul_y']= b['four wicket haul_y'].replace(to_replace='-',value='0')
b['fifer_x']= b['fifer_x'].replace(to_replace='-',value='0')
b['fifer_y']= b['fifer_y'].replace(to_replace='-',value='0')
```

```python
b= b.fillna(0)
b['fifer_x'] = b['fifer_x'].astype(int)
b['fifer_y'] = b['fifer_y'].astype(int)
b['four wicket haul_x'] = b['four wicket haul_x'].astype(int)
b['four wicket haul_y'] = b['four wicket haul_y'].astype(int)
b['overs_x']=b['overs_x'].astype(int)
b['overs_y']=b['overs_y'].astype(int)
b['maidens_x']= b['maidens_x'].astype(int)
b['maidens_y']=b['maidens_y'].astype(int)
# b['BBI_x']= b['BBI_x'].astype(int)
# b['BBI_y']=b['BBI_y'].astype(int)
```

```python
if '2022-2022' in b['span_y'].values:
    
    b.loc[b['span_y'] == '2022-2022', 'span_x'] = b['span_x'].replace('2021-2021', '2021-2022')
b['matches'] = b['matches_x'] + b['matches_y']
b['innings'] = b['innings_x'] + b['innings_y']
b['balls']= b['balls_x']+b['balls_y']
b['overs']= b['overs_x']+b['overs_y']
b['maidens']= b['maidens_x']+b['maidens_y']
b['runs']= b['runs_x']+b['runs_y']
b['wickets']=b['wickets_x']+ b['wickets_y']
b['average'] = b['runs']/b['wickets']
b['economy']=b['runs']/b['overs']
b['strikerate']= b['balls']/ b['wickets']
b['four wicket haul']=b['four wicket haul_x']+b['four wicket haul_y']
b['fifer']=b['fifer_x']+b['fifer_y']

b.rename(columns = {'span_x':'span'}, inplace = True)
b['span'] = b['span'].mask(b['span'] == 0, b['span_y'])
b.head()
```

```python
b.rename(columns = {'BBM_x':'BBM_2021'}, inplace = True)
b.rename(columns = {'BBM_y':'BBM_2022'}, inplace = True)
```

```python
b=b.drop(['BBI_x','BBI_y','matches_x','economy_y','innings_x','innings_y','balls_x','balls_y','maidens_x','runs_x','overs_y','strike rate_x','strike rate_y','wickets_y','wickets_x','runs_y','average_x','average_y' ,'fifer_x','fifer_y','four wicket haul_x','four wicket haul_y','economy_x','economy_y','overs_x','matches_y','maidens_y','span_y' ],axis=1)
```

```python
b = b.sort_values(by='wickets', ascending=False)
cols=['Player',	'span',	'matches',	'innings',	'balls','overs','maidens','runs','wickets','BBM_2021','BBM_2022','average','economy','strikerate','four wicket haul','fifer']
b=b[cols]
b.to_csv('bowling_records.csv', index=False)
b
```
